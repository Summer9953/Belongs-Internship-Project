```python
import os
import pandas as pd

folder_path = r'C:\Users\ahsan\OneDrive\Desktop\B Project\Pro\Converted file'
# Get a list of CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

dfs = []  # Create an empty list to store dataframes

# Iterate over each file
for file in csv_files:
    try:
        file_path = os.path.join(folder_path, file)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        dfs.append(df)  # Append the dataframe to the list
        
        # Output the column names
        # print(f"Columns of '{file}':")
        
        # print(df.columns.tolist())
        # print()
    except Exception as E:
        print(file_path) 
        pass   

# Now iterate over the dataframes in dfs
for df in dfs:
    missing_values = df.isnull().sum()
    if missing_values.sum() > 0:
        print("Missing values in DataFrame:")
        print(f'missing_values:{missing_values}')
    
    duplicate_rows = df[df.duplicated()]
    if not duplicate_rows.empty:
        print("Duplicate records in DataFrame:")
        print(duplicate_rows)
```
